```python
import pandas as pd
```

```python
data = pd.read_csv('../data/preprocessed/train_data.csv')
```

```python
data = data.sort_values(by=['season', 'date', 'stage'])
```

```python
from statsmodels.tsa.stattools import adfuller


def adf_test(series, signif=0.05):
    result = adfuller(series.dropna(), autolag='AIC')
    return pd.Series({
        'adf_stat': result[0],
        'p_value': result[1],
        'is_stationary': result[1] < signif
    })


adf_results = (data.drop(['match_api_id', 'season', 'stage', 'date',
                          'result_match', 'seasonal_context', 'day_of_week', 'is_weekend',], axis=1)
               .apply(adf_test))

adf_results = adf_results.T
adf_results
```
